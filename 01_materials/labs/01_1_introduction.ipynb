# Application Programming Interfaces

An Application Programming Interface (API) is a mechanism that enables two software components to communicate with each other using a set of definitions and protocols. ([AWS](https://aws.amazon.com/what-is/api/))

+ An *application* is any software with a distinct function. 
+ An *interface* can be seen as a contract between two applications that specifies how they will communicate with each other.

![](./img/01_api.svg)

We will be doing something slightly different from this diagram - we will communicate with a gateway that's situated between client and API. Rather than connect directly to the API from the client, we will be communicating with the gateway. 

This is the standard definition of an API gateway: 
+ an API gateway is a single entry point that intercepts client requests for access to backend services. 
+ the gateway delegates client requests to the correct backend microservice; e.g. if a client submits a request to "book an appointment with a doctor", the gateway would distribute tasks to an appointment scheduling microservice, a payment processor microservice, a record repository microservice, and a messaging microservice (to send email/text receipts & reminders to the user) 
+ the gateway helps with rate limiting (prevent backend from being overwhelmed by user requests), load balancing (distribute traffic across multiple instances of the same backend service), and security (security protocols, authentication, shield internal microservice URLs from the client)

In this case, what he's talking about is a second gateway that's on the client side rather than the API's side: it's U of T's gateway 
+ this U of T gateway would be communicating with the API gateway 
+ rather than managing backend services, its security functions are emphasized, and it manages communications between users of U of T's OpenAI license and OpenAI 
+ e.g. if my API key gets stolen, U of T can deactivate it and give me a new one 

## Types of APIs

There are four ways an API can work:

### REST APIs

+ Representational State Transfer.
+ Most popular and flexible APIs found today.
+ Client sends request to the server as data.
+ Server uses this client input to start internal functions and returns output data back to the client.
+ Defines a set of functions like GET, PUT, DELETE, etc. that clients can use to access server data.
+ Clients and servers exchange data using HTTP.
+ REST APIs are stateless: servers do not save client data between requests. 

### SOAP APIs

+ Simple Object Access Protocol. 
+ Client and server exchange using XML. 
+ Popular in the past, but less flexible than more modern alternatives.

### RPC APIs

+ Remote Procedural Calls.
+ Client completes a function (or procedure) on the server. 
+ The server sends the output back to the client.

### Websocket APIs

+ Uses JSON objects to pass data.
+ Supports two-way communication between client app and server.
+ Server can send callback messages to connected clients, making it more efficient than REST API.

just like Python and R are two different coding languages that can be used to code for the same things (think seaborn and ggplot), these 4 API types are different protocols that can be used to interface between softwares. 

## API Endpoints

API endpoints are the final touchpoints in the API communication system. API endpoints can be server URLs, services, and other specific digital locations where information is sent and received between systems.

Two critical aspects about API Endpoints are:

1. Security: API endpoints make the system vulnerable to attack.
2. Performance: API endpoints, particularly high-traffic ones, can cause bottlenecks.

Endpoint ≠ gateway. One application architecture usually has many endpoints but only one gateway. As an example, the gateway would be the security guard of an office building, whereas endpoints are specific office numbers or desks inside the office.

# Authentication

+ Authenticating to the OpenAI service is done through an SSH Key, Secret Key, or API Key.

## Obtaining and Using API Keys

# Hello World!

We will use the [OpenAI Python API library](https://github.com/openai/openai-python?tab=readme-ov-file) as our main tool to communicate with OpenAI's API.

The code cell below makes a first call to the Responses API. 

Note that api_key is not API_GATEWAY_KEY; API key is an alphanumeric string that identifies a client application to the API (i.e. allows OpenAI to identify me). So it is set to 'any value' as I don't care what identifier I'm given 

````python
from openai import OpenAI #importing the OpenAI client 
import os

#retrieve env values
USE_GATEWAY = (os.getenv('USE_GATEWAY', 'FALSE').upper() == 'TRUE')
MODEL = os.getenv('MODEL', 'gpt-4o-mini')

def get_client(use_gateway: bool = USE_GATEWAY) -> OpenAI: #create a client
    if use_gateway:
        client = OpenAI(base_url='https://k7uffyg03f.execute-api.us-east-1.amazonaws.com/prod/openai/v1', #open endpoint to this url so that it can send & receive messages from the url
                    api_key='any value', 
                    default_headers={"x-api-key": os.getenv('API_GATEWAY_KEY')})
    else:
        client = OpenAI()
    return client

client = get_client()

response = client.responses.create(
    model = MODEL,
    input = 'Hello world!' #sending this input to the gpt-4o-mini model and saving the output from the model to response 
    
)

print(response.output_text)
````

````python
print(response.output_text)
````

>
> Some of our official SDKs include an output_text property on model responses for convenience, which aggregates all text outputs from the model into a single string. This may be useful as a shortcut to access text output from the model.

Finally, we show the JSON-serialized version of the response object. The response object offers two methods to obtain JSON and dictionary versions of the repsonse: `response.to_json()` and `response.model_dump()`.
