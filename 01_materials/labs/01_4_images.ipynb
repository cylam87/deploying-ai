Encode the image file.

![](./img/00_cat.jpg)

```python
import base64
# Function to encode the image
def encode_image(image_path): #encode a binary object into text
    with open(image_path, "rb") as image_file: #rb stands for "read a binary object"
        return base64.b64encode(image_file.read()).decode("utf-8") #utf-8 is the encoding method, converts text into binary 
# Path to your image
image_path = "./img/00_cat.jpg"

# Getting the Base64 string
base64_image = encode_image(image_path)


response = client.responses.create(
    model=MODEL,
    input=[
        {
            "role": "user",
            "content": [
                { "type": "input_text", "text": "what's in this image?" },
                {
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_image}",
                },
            ],
        }
    ],
)
```

```python
display(Markdown(response.output_text))
```
